```python
pd.get_dummies(benefits.type_benefit).shape
```

```python
benefits = pd.concat([benefits, pd.get_dummies(benefits.type_benefit)], axis=1)
```

```python
del benefits['type_benefit']
```

```python
benefits.head()
```

```python
new_benefits = {
    'user_id': [],
    'good_job': [],
    'holiday': [],
    'motivation': []
}

for user, group in benefits.groupby('user_id'):
    new_benefits['user_id'].append(user)
    new_benefits['good_job'].append(group['good_job'].sum())
    new_benefits['holiday'].append(group['holiday'].sum())
    new_benefits['motivation'].append(group['motivation'].sum())
```

```python
new_benefits = pd.DataFrame(new_benefits)
```
